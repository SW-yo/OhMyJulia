```julia
using Pkg
Pkg.activate(@__DIR__)
using Statistics
using LinearAlgebra
using DataFrames
using CairoMakie
```

```julia
x = -3π:0.1:3π
# 内包表記の方が一般的かつシンプル
# y = [cos(sin(xi) + xi) for xi in x]
# あえてfor文で描いた場合
y = Float64[]
for xi in x
    push!(y, cos(sin(xi) + xi))
end

fig = Figure(size = (400, 400))
ax = Axis(fig[1, 1])
lines!(ax, x, y, color=:blue, label="cos(sin(x)+x)")


y = [cos(2 * atan(xi)) for xi in x]
lines!(ax, x, y, color=:red, label="cos(2*arctan(x))")

y = [(2 * exp(-xi^2) - 1) for xi in x]
lines!(ax, x, y, color=:black, label="2*exp(-x^2)-1")

axislegend(ax, position = :rt)

save("cos-sin_curve.png", fig)
fig
```
